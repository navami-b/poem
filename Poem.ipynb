```python
# poem_pipeline.py
from transformers import pipeline

prompt = "Write a small poem about the ocean"

# Create a text-generation pipeline (uses model="gpt2" by default if model not given)
generator = pipeline("text-generation", model="gpt2")  # replace with "gpt2-medium" if installed

# Generate
outputs = generator(
    prompt,
    max_length=80,        # prompt + generated tokens; adjust for length
    do_sample=True,
    temperature=0.8,
    top_p=0.9,
    num_return_sequences=1
)

# pipeline returns a list; 'generated_text' usually contains the prompt plus continuation
generated_text = outputs[0]["generated_text"]
# remove prompt if present so we print only the added poem
poem = generated_text[len(prompt):].strip() if generated_text.startswith(prompt) else generated_text

print("Poem:\n")
print(poem)
```
